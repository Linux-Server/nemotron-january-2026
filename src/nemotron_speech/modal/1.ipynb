```python
from openai import OpenAI

client = OpenAI(
    api_key="EMPTY",  # or your API key if enabled
    base_url="https://linux-server--qwen-8b-awq-vllm-serve.modal.run/v1"
)

response = client.chat.completions.create(
    model="Qwen/Qwen3-8B-AWQ",
    messages=[
        {
            "role": "user",
            "content": "Who are you?"
        }
    ],
     extra_body={
        "chat_template_kwargs": {
            "enable_thinking": False
        }
    },
    temperature=0.7,
)

print(response.choices[0].message.content)
```
